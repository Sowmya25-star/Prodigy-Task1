```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
from datetime import datetime
import warnings

# Suppress warnings
warnings.filterwarnings('ignore')

# Create synthetic traffic accident data
data = {
    'accident_time': ['2025-04-20 14:30:00', '2025-04-21 08:45:00', '2025-04-22 17:00:00', '2025-04-22 23:00:00'],
    'weather_condition': ['Clear', 'Rain', 'Fog', 'Clear'],
    'road_condition': ['Dry', 'Wet', 'Wet', 'Dry'],
    'latitude': [37.7749, 34.0522, 36.1699, 40.7128],
    'longitude': [-122.4194, -118.2437, -115.1398, -74.0060]
}

# Create DataFrame
df = pd.DataFrame(data)

# Convert accident time column to datetime
df['accident_time'] = pd.to_datetime(df['accident_time'])

# Extract hour from the accident time column
df['hour_of_day'] = df['accident_time'].dt.hour

# Clean the data (no missing values in this synthetic data)
df_clean = df

# Create visualizations

# Weather vs. Accident Frequency with distinct color
plt.figure(figsize=(10, 6))
sns.countplot(x='weather_condition', data=df_clean, palette='Blues')  # Set palette to 'Blues' for weather condition
plt.title('Accidents by Weather Condition')
plt.xticks(rotation=45)
plt.savefig('accidents_by_weather_condition.png')
plt.show()

# Time of Day vs. Accident Frequency with distinct color
plt.figure(figsize=(10, 6))
sns.countplot(x='hour_of_day', data=df_clean, palette='Set2')  # Set palette to 'Set2' for time of day
plt.title('Accidents by Hour of Day')
plt.savefig('accidents_by_hour_of_day.png')
plt.show()

# Road Condition vs. Accident Frequency with distinct color
plt.figure(figsize=(10, 6))
sns.countplot(x='road_condition', data=df_clean, palette='RdYlBu')  # Set palette to 'RdYlBu' for road condition
plt.title('Accidents by Road Condition')
plt.xticks(rotation=45)
plt.savefig('accidents_by_road_condition.png')
plt.show()

# Hotspot Visualization (Geographical)
# Assuming you want to plot the hotspots based on synthetic latitude and longitude data
gdf = gpd.GeoDataFrame(df_clean, geometry=gpd.points_from_xy(df_clean['longitude'], df_clean['latitude']))

# Plot
```
