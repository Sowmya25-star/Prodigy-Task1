```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data
data = {
    'Age': [21, 23, 25, 30, 35, 28, 40, 22, 33, 31, 26, 29, 38, 24, 27, 36],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male',
               'Male', 'Female', 'Female', 'Male', 'Female', 'Male',
               'Female', 'Male', 'Female', 'Male']
}

df = pd.DataFrame(data)
sns.set(style="whitegrid")

# Histogram of Age
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', bins=8, kde=True, color='skyblue')
plt.title('Histogram of Age', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.tight_layout()
plt.savefig('age_distribution.png')
plt.show()

# Bar chart of Gender
plt.figure(figsize=(6, 5))
sns.countplot(data=df, x='Gender', palette='pastel')
plt.title('Bar Chart of Gender', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.tight_layout()
plt.savefig('gender_distribution.png')
plt.show()

print("\nPlots displayed and saved successfully.")
print("\nAge Summary:")
print(df['Age'].describe())
print("\nGender Count:")
print(df['Gender'].value_counts())
```
