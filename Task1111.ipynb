```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data = {
    'Age': [23, 45, 25, 34, 35, 28, 50, 40, 29, 30, 33, 39, 60, 26, 32],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female', 'Female', 'Male']
}


df = pd.DataFrame(data)


plt.figure(figsize=(8, 6))


plt.subplot(2, 1, 1)  # 2 rows, 1 column, first plot
sns.histplot(df['Age'], kde=True, bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')


plt.subplot(2, 1, 2)  # 2 rows, 1 column, second plot
sns.countplot(x='Gender', hue='Gender', data=df, palette='Set2', legend=False)
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```
