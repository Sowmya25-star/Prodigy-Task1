```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
from datetime import datetime
import warnings

warnings.filterwarnings('ignore')
data = {
    'accident_time': ['2025-04-20 14:30:00', '2025-04-21 08:45:00', '2025-04-22 17:00:00', '2025-04-22 23:00:00'],
    'weather_condition': ['Clear', 'Rain', 'Fog', 'Clear'],
    'road_condition': ['Dry', 'Wet', 'Wet', 'Dry'],
    'latitude': [37.7749, 34.0522, 36.1699, 40.7128],
    'longitude': [-122.4194, -118.2437, -115.1398, -74.0060]
}

df = pd.DataFrame(data)
df['accident_time'] = pd.to_datetime(df['accident_time'])
df['hour_of_day'] = df['accident_time'].dt.hour

df_clean = df
plt.figure(figsize=(10, 6))
sns.countplot(x='weather_condition', data=df_clean, palette='Blues')
plt.title('Accidents by Weather Condition')
plt.xticks(rotation=45)
plt.savefig('accidents_by_weather_condition.png')
plt.show()
plt.figure(figsize=(10, 6))
sns.countplot(x='hour_of_day', data=df_clean, palette='Set2')
plt.title('Accidents by Hour of Day')
plt.savefig('accidents_by_hour_of_day.png')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='road_condition', data=df_clean, palette='RdYlBu')
plt.title('Accidents by Road Condition')
plt.xticks(rotation=45)
plt.savefig('accidents_by_road_condition.png')
plt.show()
gdf = gpd.GeoDataFrame(df_clean, geometry=gpd.points_from_xy(df_clean['longitude'], df_clean['latitude']))


```
